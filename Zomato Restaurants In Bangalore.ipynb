```python
import pandas as pd
import numpy as np
df=pd.read_csv('zomato.csv')
```

```python
df=df.drop_duplicates(subset ='name')
```

```python
df.index
```

```python
df=df.drop(columns=['address','phone','dish_liked','menu_item','reviews_list'])
```

```python
df['approx_cost(for two people)']=df['approx_cost(for two people)'].replace(np.nan,'NaN')
df=df[df['approx_cost(for two people)']!='NaN']
df['approx_cost(for two people)'].unique()
```

```python
df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace(',','')
df['approx_cost(for two people)']=df['approx_cost(for two people)'].dropna()
df['approx_cost(for two people)'].unique()
```

```python
df['approx_cost(for two people)']=df['approx_cost(for two people)'].astype('float')
df['price_in_dollar']=df['approx_cost(for two people)'] / 88.7
```

```python
df['rate']=df['rate'].astype('string')
df['rate']=df['rate'].str.replace('.','')
```

```python
import re
def get_numbers(plus,string):
    if plus == False:
        result=re.findall(r'\d',string)
    elif plus == True:
        result=re.findall(r'\d+',string)
    else:
        pass
    return result
for i in df['rate'].unique():
    try:
        r=get_numbers(True,i)
        df['rate']= df['rate'].str.replace(i,r[0])
    except:
        pass
df['rate'].unique()
```

```python
cbum=df['rate'].isin(['NEW','-']) | df['rate'].isna()
car=df[~cbum].copy()
car['rate']=car['rate'].astype('float')
avg=car['rate'].mean()
```

```python
df['rate']=df['rate'].replace(['NEW','-',np.nan],'37')
```

```python
df['rate']=df['rate'].astype('float')
df['rate']=df['rate'] / 10
df['rate'].unique()
```

```python
df=df.dropna(subset=['cuisines'])
df=df.dropna(subset=['rest_type'])
```

```python
df['cuisines']=df['cuisines'].astype('string')
df['cuisines']=df['cuisines'].str.split(',')
df1=df.explode('cuisines').reset_index(drop=True)
df1['cuisines']=df1['cuisines'].str.strip()
df1['cuisines'].unique()
```

```python
df['rest_type']=df['rest_type'].astype('string')
df['rest_type']=df['rest_type'].str.split(',')
df2=df.explode('rest_type').reset_index(drop=True)
df['rest_type']=df['rest_type'].replace(' Pub','Pub')
df2['rest_type']=df2['rest_type'].str.strip()
```

```python
df1.to_csv('expanded_cuisines.csv')
df2.to_csv('expanded_type.csv')
```
