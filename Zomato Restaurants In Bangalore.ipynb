```python
import pandas as pd
import numpy as np
df=pd.read_csv(r'C:\Users\ASUS\Desktop\Python Working On\zomato.csv')   ## The main dataset
```

```python
df=df.drop(columns=['address','phone','dish_liked','menu_item','reviews_list'])   ## Removing unnecessary columns
```

```python
df=df.drop_duplicates(subset ='name')   ## Most restaurants have multiple rows, these rows are now identical after removing the previous columns
```

```python
df['approx_cost(for two people)']=df['approx_cost(for two people)'].replace('NaN',np.nan)
df=df.dropna(subset=['approx_cost(for two people)'])
df['approx_cost(for two people)'].unique()
   ## Eliminating restaurants with invalid price column data
```

```python
df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace(',','')
df['approx_cost(for two people)'].unique()
   ## Removing commas before changing the data type to numeric
```

```python
df['approx_cost(for two people)']=df['approx_cost(for two people)'].astype('float')   ## Data type changed
df['price_in_dollar']=df['approx_cost(for two people)'] / 88.7   ## Converting Indian Ruppes to  USD
```

```python
df['rate'].unique()
   ## The rate column will be changed to numeric values
```

```python
df['rate']=df['rate'].astype('string')   ## Changing the data type from object to string in order to access the characters
df['rate']=df['rate'].str.replace('.','')   ## Removing the period '.' to access consecutive numbers
```

```python
import re
def get_numbers(string):
    result=re.findall(r'\d+',string)
    return result
   ## A function for finding the two numbers representing the rate
for i in df['rate'].unique():
    try:
        r=get_numbers(i)
        df['rate']= df['rate'].str.replace(i,r[0])
    except:
        pass
   ## Replacing the text with only the numbers, errors will be handled next
df['rate'].unique()
```

```python
missing_values=df['rate'].isin(['NEW','-']) | df['rate'].isna()
valid_data=df[~missing_values].copy()
   ## Selecting numeric data to replace missing values with the average
valid_data['rate']=valid_data['rate'].astype('float')
avg=valid_data['rate'].mean().astype(int)
print(avg)
## The average is 37, note the results may slightly change if the Kaggle dataset was updated
```

```python
df['rate']=df['rate'].replace(['NEW','-',np.nan],str(avg))   ## Missing values replaced
```

```python
df['rate']=df['rate'].astype('float')
df['rate']=df['rate'] / 10
df['rate'].unique()
   ## Dividing by 10, now the rate column shows the rate out of 5
```

```python
df=df.dropna(subset=['cuisines'])
df=df.dropna(subset=['rest_type'])
   ## Removing restaurants with no registed type or cuisine
```

```python
df['cuisines']=df['cuisines'].astype('string')   ## Changing the data type from object to string in order to access the characters
df['cuisines']=df['cuisines'].str.split(',')   ## Replacing commas with spaces
df1=df.explode('cuisines').reset_index(drop=True)
   ## Creating a separate dataframe containing each cuisine as a separate category to ensure an accurate count of each cuisine
   ## The original column hosted cells with multiple cuisines that would have been treated as one
df1['cuisines']=df1['cuisines'].str.strip()   ## Removing all spaces to avoid unique cuisines appearing twice
df1=df1['cuisines']   ## Only this column will be used in the report
df1.head()
```

```python
df['rest_type']=df['rest_type'].astype('string')   ## Changing the data type from object to string in order to access the characters
df['rest_type']=df['rest_type'].str.split(',')   ## Replacing commas with spaces
df2=df.explode('rest_type').reset_index(drop=True)
   ## Creating a separate dataframe containing each type as a separate category to ensure an accurate count of each type
   ## The original column hosted cells with multiple types that would have been treated as one
df2['rest_type']=df2['rest_type'].str.strip()   ## Removing all spaces to avoid unique cuisines appearing twice
df2=df2[['rest_type','name']] ## Only these columns will be used in the report
df2['rest_type']=df2['rest_type'].dropna()
df2.head()
```

```python
df.to_csv('restaurant-data.csv')
df1.to_csv('expanded_cuisines.csv')
df2.to_csv('expanded_type.csv')
   ## Dataframes saved
```
